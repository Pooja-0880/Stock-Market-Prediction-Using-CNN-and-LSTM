```python
# concatinating the datasets 
df_aapl = pd.concat([df_aapl_1,df_aapl_2,df_aapl_3,df_aapl_4,df_aapl_5,df_aapl_6,df_aapl_7,df_aapl_8,df_aapl_9,df_aapl_10],axis=0)


df_googl= pd.concat([df_googl_1,df_googl_2,df_googl_3],axis=0)

df_amzn = pd.concat([df_amzn_1,df_amzn_2,df_amzn_3], axis=0)

df_jnj = pd.concat([df_jnj_1,df_jnj_2,df_jnj_3, df_jnj_4, df_jnj_5], axis=0)

df_ma = pd.concat([df_ma_1,df_ma_2,df_ma_3],axis=0)
```

```python
def clean_columns(df):
    
     # remove extra column 
    df.drop(['Unnamed: 0','id'], axis=1, inplace = True)  
    
    # removing the time from date_time column, keeping only date
    df['date_time'] = pd.to_datetime(df['date_time'])
    df['date_time'] = df['date_time'].dt.date
    return df


df_list = [df_aapl, df_googl,df_ma, df_amzn, df_jnj]

for df in df_list: 
    df= clean_columns(df)
```
